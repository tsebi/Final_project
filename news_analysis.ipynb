```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()

def classify_text(text):
    try:
        scores = sid.polarity_scores(text)
        if scores['compound'] == 0.0:
            return 'neutral', scores['compound'], scores['neu']
        elif scores['compound'] > 0.0:
            return 'positive', scores['compound'], scores['neu']
        else:
            return 'negative', scores['compound'], scores['neu']
    except Exception as e:
        print(f"Error occurred: {str(e)}")
        return 'error', None, None
```

```python
sentiment_analysis_df['Class'], sentiment_analysis_df['Polarity'], sentiment_analysis_df['Subjectivity'] = zip(*sentiment_analysis_df['text'].apply(classify_text))
```
