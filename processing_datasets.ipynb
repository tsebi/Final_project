# Oil dataframe

```python
df_oil = pd.read_csv('/content/drive/MyDrive/Final project/oil.csv')
```

```python
df_oil.head()
```

```python
df_oil.columns
```

```python
from datetime import datetime
```

```python
dates = []
prices = []

for _, row in df_oil.iterrows():
  start_date = row['date'][0:11]
  end_date = row['date'][0:4] + ' ' + row['date'][15:]

  start_date_dt = datetime.strptime(start_date, '%Y %b-%d').date()
  end_date_dt = datetime.strptime(end_date, '%Y %b-%d').date()

  dates.extend([start_date_dt] + [start_date_dt + (end_date_dt - start_date_dt) / 4 * i for i in range(1, 4)] + [end_date_dt])

  price_values = row[['monday', 'tuesday', 'wednesday', 'thursday', 'friday']].values.tolist()
  prices.extend(price_values)
```

```python
display(len(dates), len(prices))
```

```python
print(dates)
```

```python
print(prices)
```

```python
data_tuples = list(zip(dates, prices))
df_oil_processed = pd.DataFrame(data_tuples, columns = ['Date', 'Oil'])
```

```python
display(df_oil_processed.info(), df_oil_processed.head())
```

```python
df_oil_processed['Date'] = pd.to_datetime(df_oil_processed['Date'])
```

```python
df_oil_processed = unique_dates_df.copy()
```

```python
def get_df_name(df):
    return [name for name in globals() if globals()[name] is df][0]
```

```python
def calculate_days(dataframe):
  min_date = dataframe['Date'].min()
  max_date = dataframe['Date'].max()

  days_in_df = (max_date - min_date).days + 1

  return days_in_df
```
