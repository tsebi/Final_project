```python
# Open the Google Sheets file
file = gc.open('bitcoin')

# Create an empty list to store the dataframe names
dataframe_names = []

# Create a folder in Google Drive to save the CSV files
folder_path = '/content/drive/MyDrive/Final project'
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
    print(f"Created folder: {folder_path}")

# Iterate over each worksheet in the file
for worksheet in file.worksheets():
    # Get the sheet name
    sheet_name = worksheet.title

    # Get the data from the worksheet
    data = worksheet.get_all_values()

    # Create a dataframe with sheet name as df_sheet_name
    df_name = "df_" + sheet_name.replace(' ', '_')  # Remove spaces and add prefix
    df = pd.DataFrame(data[1:], columns=data[0])

    # Store the dataframe in the list
    globals()[df_name] = df  # Assign dataframe to a dynamically named variable
    dataframe_names.append(df_name)

    # Print the dataframe
    print(f"Dataframe for sheet '{sheet_name}':")
    print(df.head())
    print("\n")

    # Save the dataframe to a CSV file in Google Drive
    csv_filename = f"{folder_path}/{sheet_name}.csv"
    df.to_csv(csv_filename, index=False)
    print(f"CSV file saved: {csv_filename}")

# Print the list of dataframe names
print("List of dataframe names:")
print(dataframe_names)
```
